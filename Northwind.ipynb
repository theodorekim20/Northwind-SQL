# Ranking Employee Sales Performance

Create a CTE that calculates the total sales for each employee. Rank the sales

```python
%%sql
WITH employee_orders AS (
    SELECT 
        e.first_name || ' ' || e.last_name AS employee_name,
        o.order_id,
        o.order_date,
        od.quantity,
        od.unit_price
    FROM orders o
    JOIN employees e ON o.employee_id = e.employee_id
    JOIN order_details od ON o.order_id = od.order_id
)
SELECT 
    employee_name,
    order_id,
    order_date,
    (quantity * unit_price) AS total_sales,
    RANK() OVER (ORDER BY (quantity * unit_price) DESC) AS sales_rank
FROM employee_orders
LIMIT 10;
```

# Running Total of Sales

Join the Orders and Order_Details tables to bring together the data you need.
Group by the month of the Order_Date and calculate the total sales for each month. Use the DATE_TRUNC function to truncate the Order_Date to the nearest month.
DATE_TRUNC- Aggregates data by a specific time unit. In this case it does it by month.
Use the SUM function with an OVER clause to calculate the running total of sales by month.

```python
%%sql
WITH orders_orderdetails AS (
    SELECT 
        o.order_id,
        od.quantity,
        od.unit_price,
        DATE_TRUNC('month', o.order_date) AS order_month
    FROM orders o
    JOIN order_details od ON o.order_id = od.order_id
)
SELECT 
    order_id,
    order_month,
    (quantity * unit_price) AS total_sales,
    SUM(quantity * unit_price) OVER (PARTITION BY order_month) AS running_sales_month
FROM orders_orderdetails
LIMIT 10;
```

# Month Over Month Sales Growth

Create a CTE calculating total sales for each month
Create second CTE using Lag function with OVer clause to get total sales of previous Month
Main query calculate month-over-month sales growth rate.

```python
%%sql
WITH monthly_sales AS (
    SELECT 
        DATE_TRUNC('month', o.order_date) AS order_month,
        SUM(od.quantity * od.unit_price) AS total_sales
    FROM orders o
    JOIN order_details od ON o.order_id = od.order_id
    GROUP BY DATE_TRUNC('month', o.order_date)
),
sales_with_previous AS (
    SELECT 
        order_month,
        total_sales,
        LAG(total_sales) OVER (ORDER BY order_month) AS previous_month_sales
    FROM monthly_sales
)
SELECT 
    order_month,
    total_sales,
    previous_month_sales,
    CASE 
        WHEN previous_month_sales IS NOT NULL THEN 
            ROUND((total_sales::NUMERIC - previous_month_sales::NUMERIC) / previous_month_sales::NUMERIC * 100, 2)
        ELSE 
            NULL
    END AS month_over_month_growth_rate
FROM sales_with_previous
ORDER BY order_month;

```

# Identifying High-Value Customers

Create a CTE including customer identification and calculates value of each of their orders
In main query use CTE to categorize each order as 'Above Average' or 'Average/Below Average' using a CASE statement.

```python
%%sql
WITH customer_sales AS (
    SELECT 
        c.customer_id,
        SUM(od.quantity * od.unit_price) AS total_customer_sales
    FROM customers c
    JOIN orders o ON o.customer_id = c.customer_id
    JOIN order_details od ON o.order_id = od.order_id
    GROUP BY c.customer_id
),
average_sales AS (
    SELECT 
        AVG(total_customer_sales) AS avg_sales
    FROM customer_sales
)
SELECT 
    cs.customer_id,
    cs.total_customer_sales,
    CASE 
        WHEN total_customer_sales > AVG(total_customer_sales) OVER () THEN 
            'Above Average'
        ELSE 
            'Average/Below Average'
    END AS sales_category
FROM customer_sales cs
ORDER BY cs.total_customer_sales;
```

# Percentage of Sales for Each Category

Find the percentage of total sales for each product category.

Create a CTE that calculates the total sales for each product category.
Use your CTE in the main query to calculate the percentage of total sales for each product category.


```python
%%sql
WITH CategorySales AS (
    SELECT Categories.Category_ID, Categories.Category_Name,
           SUM(Products.Unit_Price * Quantity * (1 - Discount)) AS "Total Sales"
    FROM Categories
    JOIN Products ON Categories.Category_ID = Products.Category_ID
    JOIN Order_Details ON Products.Product_ID = Order_Details.Product_ID
    GROUP BY Categories.Category_ID
)
SELECT Category_ID, Category_Name,
       "Total Sales" / SUM("Total Sales") OVER () * 100 AS "Sales Percentage"
FROM CategorySales;
```

# Finding the top 3 products sold in each category

```python
%%sql
WITH ProductSales AS (
    SELECT Products.Category_ID, 
           Products.Product_ID, Products.Product_Name,
           SUM(Products.Unit_Price * Quantity * (1 - Discount)) AS "Total Sales"
    FROM Products
    JOIN Order_Details ON Products.Product_ID = Order_Details.Product_ID
    GROUP BY Products.Category_ID, Products.Product_ID
)
SELECT Category_ID, 
       Product_ID, Product_Name,
       "Total Sales"
FROM (
    SELECT Category_ID, 
           Product_ID, Product_Name,
           "Total Sales", 
           ROW_NUMBER() OVER (PARTITION BY Category_ID ORDER BY "Total Sales" DESC) AS rn
    FROM ProductSales
) tmp
WHERE rn <= 3;
```
